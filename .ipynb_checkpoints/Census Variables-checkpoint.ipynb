### import libraries

```python
import pandas as pd
import numpy as np
import datetime
```

```python
# Read in the census dataframe
census = pd.read_csv('census_data.csv', index_col=0)
#checking the data
print(census.head())
```

```python
#checking data types
print(census.dtypes)
# print(census.info())
```

```python
#checking the average year of respondents
#change birth year data type for easy computation
#change missing values to nan
# print(census.birth_year.unique())
census.birth_year = census.birth_year.replace("missing",1967)
# print(census.birth_year.unique())
#change birth year data type for easy computation
census.birth_year = census.birth_year.astype("int")
#checking mean year
mean_birth_year = int(census.birth_year.mean())

print(mean_birth_year)
```

```python
# making tax ordinal data
census["higher_tax"] = pd.Categorical(census["higher_tax"],["strongly disagree" , "disagree" , "neutral" , "agree" , "strongly agree"],ordered = True)
print(census["higher_tax"].unique())
# label encode higher tax to find median
census["higher_tax_codes"] = census["higher_tax"].cat.codes
print(census["higher_tax_codes"].median())
# code label marital status
print(census["marital_status"].unique())
census["marital_status"] =pd.Categorical(census["marital_status"],['single' ,'divorced' ,'married' ,'widowed'],ordered = False)
census["marital_codes"] = census["marital_status"].cat.codes
print(census)
#One-Hot encode marital status
census = pd.get_dummies(data = census , columns =["marital_status"])

print(census)
```

```python
# get age of respondents
now = datetime.datetime.now()
age_f = lambda x: now.year - x
census["age"] = census["birth_year"].apply(age_f)
print(census)
bins= list(range(10,90,5))
print(bins)

census["age_group"] = pd.cut(census.age,bins)
print(census)
census["age_group_labels"] = census["age_group"].cat.codes
print(census["age_group_labels"].unique())
```
